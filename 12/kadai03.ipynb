```python
def solve_difequation(difequ,initlis,tstart,tend,dt = 0.01):
    """difequ:微分方程式のリスト(リストのi番目の要素が未知関数のi番目の変数をtについて微分した式)
          initlis:リストのi番目の要素が未知関数のi番目の変数のt=tstartにおける値
          tstart:パラメータtの開始値
          tend:パラメータtの終了値
          dt:オイラー法におけるデルタt"""
    dim = len(difequ)#未知関数の変数の数
    t = tstart
    ans = [[tstart,*initlis]]
    t+=dt
    while t < tend:
        tnextlis=[t]
        for i in range(dim):
            tnextlis.append( ans[-1][i+1] + dt*difequ[i](*(ans[-1][1:])) )
        ans.append(tnextlis[:])
        t+=dt
    return ans
```

```python
fs = [lambda x,y: x - x**3 - y, lambda x,y: x ]
traj = solve_difequation(fs, [2,1], 0, 30)
```

```python
import matplotlib.pyplot as plt
ttr,xtr,ytr = zip(*traj)
plt.plot(xtr,ytr)
plt.show()
```

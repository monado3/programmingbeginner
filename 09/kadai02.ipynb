```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def logistic(a,x0):
    if a<0 or 4<a:
        raise ValueError("'a' needs to be 0 <= a <= 4")
    elif x0<0 or 1<x0:
        raise ValueError("'x0' needs to be 0 <= x0 <= 1")
    else:
        x=x0
        while True:
            xnext=a*x*(1.0-x)
            yield xnext
            x=xnext
```

```python
def plot(x0=0.1,n_max=1000,n_min=800,astep=300):
    xlis=[]  
    ylis=[]
    for i in range(astep):
        a=4.0 * i /astep
        l=logistic(a,x0)
        for j in range(n_max):
            g=next(l)
            if j >= n_min:
                xlis.append(a)
                ylis.append(g)
    plt.scatter(xlis,ylis,s=0.1) 
```

```python
plot(0.23,500,100,300)
```

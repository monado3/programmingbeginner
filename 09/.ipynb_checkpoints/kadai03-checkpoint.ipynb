```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def randomwalk(x0,p):
    if not 0<=p<=1:
        raise ValueError("'p' needs to be 0 <= p <= 1")
    elif type(x0)!= int:
        raise TypeError("'x0' needs to be int.")
    else:
        x=x0
        while True:
            yield x
            if random.random() < p:
                x+=1
            else:
                x-=1
```

```python
def plot(x0=0,p=0.5,steps=1000):
    if type(steps)!= int:
        raise TypeError("'steps' needs to be int.")
    else:
        r=randomwalk(x0,p)
        xlis=[i for i in range(steps)]
        ylis=[next(r) for i in range(steps)]
        plt.scatter(xlis,ylis,s=0.4)
```

```python
plot(x0=0,p=0.5,steps=1000)
```

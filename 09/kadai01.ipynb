```python
def logistic(a,x0):
    if a<0 or 4<a:
        raise ValueError("'a' needs to be 0 <= a <= 4")
    elif x0<0 or 1<x0:
        raise ValueError("'x0' needs to be 0 <= x0 <= 1")
    else:
        x=x0
        while True:
            xnext=a*x*(1.0-x)
            yield xnext
            x=xnext
```

```python
g = logistic(a=3.9, x0=0.1)
traj = [next(g) for i in range(100)]

import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(traj)
```

```python
import mylib.ICA as ica
import mylib.IO as io
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#in order to immediately reflect change of mylib/*.py on this kadai1.ipynb.
%reload_ext autoreload
%autoreload 2           
```

```python
# load the data from files
dat1 = np.loadtxt('report2_all/report2_kadai1_data1/dat1.txt', delimiter= '\n')
dat2 = np.loadtxt('report2_all/report2_kadai1_data1/dat2.txt', delimiter= '\n')
dataX = np.array((dat1,dat2))

#linear conversion of every dataX[i] so that E[dataX[i]] = 0
dataX, meanarr = ica.make_mean_0(dataX)
```

```python
# show dataX
io.plot_graph(dataX[0], 'dat1', 'time')
plt.show()

io.plot_graph(dataX[1], 'dat2', 'time')
plt.show()
```

```python
# whiten dataX
dataZ = ica.whitening_data(dataX)

# calculate a matrix W
mat_W = ica.make_W(dataZ, 2)
```

```python
# find two source signals
source = ica.find_source(mat_W, dataZ)
```

```python
# show source
x_axis = np.arange(len(source[0]))
plt.plot(x_axis, source[0], label = 'source1')
plt.plot(x_axis, source[1], label = 'source2')
plt.title('source1 and source2')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.legend(loc = 'upper right')
plt.grid()
plt.show()
```

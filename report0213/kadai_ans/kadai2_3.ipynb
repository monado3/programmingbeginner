```python
import mylib.ICA as ica
import mylib.IO as io
import matplotlib.pyplot as plt
import scipy.io.wavfile as spwav
import numpy as np
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
# load the data from files
sound1 = spwav.read('report2_all/report2_kadai2_data3/music1.wav')[1]
sound2 = spwav.read('report2_all/report2_kadai2_data3/music2.wav')[1]
dataX = np.array((sound1,sound2))

#linear conversion of every dataX[i] so that E[dataX[i]] = 0
dataX, meanarr = ica.make_mean_0(dataX)
print(len(dataX[0]),len(dataX[1]))
```

```python
# show sound1
io.plot_graph(dataX[0], 'sound1', 'time')
plt.show()

# show sound2
io.plot_graph(dataX[1], 'sound2', 'time')
plt.show()
```

```python
# whiten dataX
dataZ = ica.whitening_data(dataX)

# calculate a matrix W
mat_W = ica.make_W(dataZ, 2)
```

```python
# calculate two source signals
source = ica.find_source(mat_W, dataZ)

# standardize each source
source = io.standardize_wav(source)
```

```python
# show source1
io.plot_graph(source[0], 'Turkischer Marsch', 'time')
plt.show()

# show source2
io.plot_graph(source[1], 'Fr Elise', 'time')
plt.show()
```

```python
# write each source to a wavfile
spwav.write('separated_source/kadai2/kadai2_3/Turkischer_Marsch.wav', 44100, source[0])
spwav.write('separated_source/kadai2/kadai2_3/Fur_Elise.wav', 44100, source[1])
```
